##BioGPT

```python
# File: evaluate_biogpt_qa.py

# Install required packages
!pip install -q transformers sentence-transformers rouge-score tqdm sacremoses nltk bert-score

import json
import pandas as pd
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
from sentence_transformers import SentenceTransformer, util
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.tokenize import word_tokenize
from bert_score import score as bertscore
import nltk
nltk.download('punkt')

# Load dataset
with open("/kaggle/working/fused_alpaca.json", "r") as f:
    data = [json.loads(line) for line in f]

df = pd.DataFrame(data).dropna(subset=["instruction", "output"]).reset_index(drop=True)
df = df[['instruction', 'output']].rename(columns={"instruction": "question", "output": "true_answer"})
df = df.sample(30, random_state=42).reset_index(drop=True)

# Load BioGPT
biogpt_name = "microsoft/BioGPT-Large"
tokenizer = AutoTokenizer.from_pretrained(biogpt_name)
model = AutoModelForCausalLM.from_pretrained(biogpt_name).to("cuda" if torch.cuda.is_available() else "cpu")
model.eval()

# Answer generation
def generate_answers_biogpt_batch(questions, max_new_tokens=100, batch_size=8):
    answers = []
    device = model.device
    for i in tqdm(range(0, len(questions), batch_size), desc="Generating answers"):
        batch = questions[i:i+batch_size]
        inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True).to(device)
        outputs = model.generate(
            **inputs,
            max_length=inputs["input_ids"].shape[1] + max_new_tokens,
            do_sample=True,
            top_p=0.9,
            temperature=0.7,
            pad_token_id=tokenizer.eos_token_id,
        )
        decoded = [tokenizer.decode(output, skip_special_tokens=True) for output in outputs]
        decoded_clean = [d.replace(q, "").strip() for d, q in zip(decoded, batch)]
        answers.extend(decoded_clean)
    return answers

df['biogpt_answer'] = generate_answers_biogpt_batch(df['question'].tolist(), batch_size=8)

# Cosine similarity using SBERT
sbert = SentenceTransformer("all-MiniLM-L6-v2")
true_embs = sbert.encode(df['true_answer'].tolist(), convert_to_tensor=True)
pred_embs = sbert.encode(df['biogpt_answer'].tolist(), convert_to_tensor=True)
cosine_scores = util.cos_sim(true_embs, pred_embs).diagonal().cpu().numpy()
df['cosine'] = cosine_scores

# ROUGE-L score
scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
tqdm.pandas(desc="Calculating ROUGE-L")
df['rougeL'] = df.progress_apply(lambda row: scorer.score(row['true_answer'], row['biogpt_answer'])['rougeL'].fmeasure, axis=1)

# BLEU Score
smooth = SmoothingFunction().method4
df['bleu'] = df.progress_apply(lambda row: sentence_bleu(
    [word_tokenize(row['true_answer'])],
    word_tokenize(row['biogpt_answer']),
    smoothing_function=smooth
), axis=1)

# BERTScore
P, R, F1 = bertscore(df['biogpt_answer'].tolist(), df['true_answer'].tolist(), lang="en")
df['bertscore_precision'] = P.tolist()
df['bertscore_recall'] = R.tolist()
df['bertscore_f1'] = F1.tolist()

# Print averages
print("Average Cosine Similarity:", df['cosine'].mean())
print("Average ROUGE-L:", df['rougeL'].mean())
print("Average BLEU:", df['bleu'].mean())
print("Average BERTScore F1:", df['bertscore_f1'].mean())

# Save
df.to_csv("biogpt_medical_qa_results_sample.csv", index=False)

# Show sample rows
print(df[['question', 'true_answer', 'biogpt_answer', 'cosine', 'rougeL', 'bleu', 'bertscore_f1']].head(5))
```

```python
# Install necessary packages
!pip install -q transformers sentence-transformers rouge-score tqdm sacremoses nltk bert-score

import json
import pandas as pd
from tqdm import tqdm
from transformers import AutoTokenizer, AutoModelForCausalLM
import torch
from sentence_transformers import SentenceTransformer, util
from rouge_score import rouge_scorer
from nltk.translate.bleu_score import sentence_bleu, SmoothingFunction
from nltk.tokenize import word_tokenize
from bert_score import score as bertscore
import nltk
import time

# NLTK setup
nltk.download('punkt')

# Check if CUDA GPU is available
print("CUDA available:", torch.cuda.is_available())
if torch.cuda.is_available():
    print("Device name:", torch.cuda.get_device_name(torch.cuda.current_device()))

# Load JSONL dataset
with open("/kaggle/working/fused_alpaca.json", "r") as f:
    data = [json.loads(line) for line in f]

df = pd.DataFrame(data).dropna(subset=["instruction", "output"]).reset_index(drop=True)
df = df[['instruction', 'output']].rename(columns={"instruction": "question", "output": "true_answer"})
df = df.sample(30, random_state=42).reset_index(drop=True)

# Load BioMedLM model and tokenizer
biomedlm_name = "stanford-crfm/BioMedLM"
tokenizer = AutoTokenizer.from_pretrained(biomedlm_name)
tokenizer.pad_token = tokenizer.eos_token
model = AutoModelForCausalLM.from_pretrained(biomedlm_name).to("cuda" if torch.cuda.is_available() else "cpu")
model.eval()

# Batch answer generation
def generate_answers_biomedlm_batch(questions, max_new_tokens=30, batch_size=2):
    answers = []
    device = model.device
    for i in tqdm(range(0, len(questions), batch_size), desc="Generating answers (BioMedLM)"):
        batch = questions[i:i+batch_size]
        start = time.time()
        inputs = tokenizer(batch, return_tensors="pt", padding=True, truncation=True).to(device)
        outputs = model.generate(
            **inputs,
            max_length=inputs["input_ids"].shape[1] + max_new_tokens,
            do_sample=True,
            top_p=0.9,
            temperature=0.7,
            pad_token_id=tokenizer.pad_token_id,
        )
        decoded = [tokenizer.decode(output, skip_special_tokens=True) for output in outputs]
        decoded_clean = [d.replace(q, "").strip() for d, q in zip(decoded, batch)]
        answers.extend(decoded_clean)
        end = time.time()
        print(f"Batch {i//batch_size + 1} generated in {end - start:.2f} seconds")
    return answers

df['biomedlm_answer'] = generate_answers_biomedlm_batch(df['question'].tolist(), max_new_tokens=30, batch_size=2)

# Cosine similarity (SBERT)
sbert = SentenceTransformer("all-MiniLM-L6-v2")
true_embs = sbert.encode(df['true_answer'].tolist(), convert_to_tensor=True)
pred_embs = sbert.encode(df['biomedlm_answer'].tolist(), convert_to_tensor=True)
cosine_scores = util.cos_sim(true_embs, pred_embs).diagonal().cpu().numpy()
df['cosine'] = cosine_scores

# ROUGE-L evaluation
scorer = rouge_scorer.RougeScorer(['rougeL'], use_stemmer=True)
tqdm.pandas(desc="Calculating ROUGE-L")
df['rougeL'] = df.progress_apply(lambda row: scorer.score(row['true_answer'], row['biomedlm_answer'])['rougeL'].fmeasure, axis=1)

# BLEU score
smooth = SmoothingFunction().method4
df['bleu'] = df.progress_apply(
    lambda row: sentence_bleu(
        [word_tokenize(row['true_answer'])],
        word_tokenize(row['biomedlm_answer']),
        smoothing_function=smooth
    ),
    axis=1
)

# BERTScore
P, R, F1 = bertscore(df['biomedlm_answer'].tolist(), df['true_answer'].tolist(), lang="en")
df['bertscore_precision'] = P.tolist()
df['bertscore_recall'] = R.tolist()
df['bertscore_f1'] = F1.tolist()

# Summary statistics
print("Average Cosine Similarity:", df['cosine'].mean())
print("Average ROUGE-L:", df['rougeL'].mean())
print("Average BLEU:", df['bleu'].mean())
print("Average BERTScore F1:", df['bertscore_f1'].mean())

# Save full result
df.to_csv("biomedlm_medical_qa_results_sample.csv", index=False)

# Show examples
print(df[['question', 'true_answer', 'biomedlm_answer', 'cosine', 'rougeL', 'bleu', 'bertscore_f1']].head(5))
```
